```python
def ler_ficheiro (ficheiro):
    abrir = open (ficheiro, "r")
    linhas = abrir.readlines ()
    linhas = linhas [1:]
    for linha in linhas:
        campos = linha.strip().split(",")
        abrir.close ()
    return linhas

#ler_ficheiro ("diabetes_prediction_dataset.csv")




```

```python
def calcular_distribuição (ficheiro):
    ler = ler_ficheiro (ficheiro)
    femdoe = 0
    femsau = 0
    masdoe = 0
    massau = 0
    for linha in ler:
        campos = linha.strip().split(",")
        genero = campos[0]
        diabetes = campos[-1]
        if genero == "Female":
            if diabetes == "0":
                femsau = femsau + 1
            if diabetes == "1":
                femdoe = femdoe + 1
        if genero == "Male":
            if diabetes == "0":
                massau = massau + 1
            if diabetes == "1":
                masdoe = masdoe + 1
    totalf = femdoe + femsau
    totalm = masdoe + massau
    pfem = float ((femdoe/totalf)*100)
    pmas = float ((masdoe/totalm)*100)
    print ("Percentagem de Mulheres Doentes = " + str (pfem) + "%")
    print ("Percentagem de Homens Doentes = " + str (pmas) + "%")


calcular_distribuição ("diabetes_prediction_dataset.csv")

def distribuição_etaria (ficheiro):
    ler = ler_ficheiro (ficheiro)
    a010a = 0
    a1019a = 0
    a2029a = 0
    a3039a = 0
    a4049a = 0
    a5059a = 0
    a6069a = 0
    a7079a = 0
    a80mais = 0
    total = 0
    for linha in ler:
        campos = linha.strip().split(",")
        idade = float (campos [1])
        diabetes = campos[-1]
        if diabetes == "1":
            total = total + 1
            if idade < 10 :
                a010a = a010a + 1
            if idade >= 80 :
                a80mais = a80mais + 1
            if idade >= 10 and idade < 20:
                a1019a = a1019a + 1
            if idade >= 20 and idade <30:
                a2029a = a2029a + 1
            if idade >= 30 and idade < 40:
                a3039a = a3039a + 1
            if idade >= 40 and idade < 50:
                a4049a = a4049a + 1
            if idade >= 50 and idade < 60:
                a5059a = a5059a + 1
            if idade >=60 and idade < 70:
                a6069a = a6069a + 1
            if idade >70 and idade <80:
                a7079a = a7079a + 1
    p010 = ((a010a/total)*100)
    p1019 = ((a1019a/total)*100)
    p2029 = ((a2029a/total)*100)
    p3039 = ((a3039a/total)*100)
    p4049 = ((a4049a/total)*100)
    p5059 = ((a5059a/total)*100)
    p6069 = ((a6069a/total)*100)
    p7079 = ((a7079a/total)*100)
    p80mais = ((a80mais/total)*100)
    print ("[0-10]: " + str (p010) + "%")
    print ("[10-19]: " + str (p1019) + "%")
    print ("[20-29]: " + str (p2029) + "%")
    print ("[30-39]: " + str (p3039) + "%")
    print ("[40-49]: " + str (p4049) + "%")
    print ("[50-59]: " + str (p5059) + "%")
    print ("[60-69]: " + str (p6069) + "%")
    print ("[70-79]: " + str (p7079) + "%")
    print ("[80+]: " + str (p80mais) + "%")

distribuição_etaria ("diabetes_prediction_dataset.csv")

# def criar_tabela (ficheiro):
    ler = ler_ficheiro (ficheiro)
    tabela = []
    for linha in ler:
        tabela.append (linha)
        print (tabela)

# criar_tabela ("diabetes_prediction_dataset.csv")

```
