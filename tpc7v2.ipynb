```python
# Aquecimento

pol1 = [(1,7), (-3.7, 4), (5,3), (-88,0)]
pol2 = [(2,2), (17,0)]
pol3 = [(7,4), (6,3), (-1,2), (43,0)]
pol4 = [(1,4)]
pol5 = [(1,2)]
pol6 = [(1,3)]

def guardar_polinomios(lista,file):
    abrir = open(file, "w")
    for pol in lista:
        for coef, grau in pol:
            abrir.write (str(coef) + "," + str(grau) + " ")
        abrir.write ("\n")
    abrir.close ()

lista = [pol1, pol2, pol3, pol4, pol5, pol6]
guardar_polinomios (lista, "polinomiostpc7.txt")
```

```python
# Desenvolvimento funções não trabalhadas em aula

# Derivada de um polinómio

def calc_derivada (lista):
    res = []
    for pol in lista:
        for coef, grau in pol:
            coef = coef * grau
            grau = grau - 1
            res.append ((coef, grau))
    return res

calc_derivada (lista)
```

```python
# Aplicação para manipular polinómios - Funções Utilizadas

def criar_polinomios ():
    grau = int (input ("Escolha o grau do termo."))
    res = []
    while grau >= 0:
        coef = float (input ("Escolha o coeficiente do termo."))
        if coef != 0:
            res.append ((coef, grau))
        grau = grau - 1
    return res

def ler_polinomios (file):
    abrir = open (file)
    leitura = []
    for linha in abrir:
        linha = linha.strip()
        termos = linha.split(" ")
        leitura.append (termos)
    abrir.close ()
    return leitura

def listar_polinomios (file):
    abrir = open (file)
    leitura = []
    n = 1
    for linha in abrir:
        linha = linha.strip()
        termos = linha.split(" ")
        termos = str (n) + "-" + str (termos)
        n = n + 1
        leitura.append (termos)
    return leitura   

def calc_polinomio (pol, valor):
    res = 0
    for coef, grau in pol:
        res = res + coef * (valor ** grau)
    return res

def listar_pol_grau (file):
    abrir = open (file)
    leitura = []
    n = 1
    for linha in abrir:
        linha = linha.strip()
        termos = linha.split(" ")
        grau = termos[0][2]
        termos = str(n) + "-" + str (termos)
        termos = termos + "; grau: " + str (grau)
        n = n + 1
        leitura.append (termos)
    return leitura
```
