```python
#Lista Funções TPC 5

def listar (cinema):
    print ("-------Em exibição-------")
    for sala in cinema:
        print ("Filme:", sala [2],"/ " "Lugares:", sala [0])
    return cinema

def disponivel (cinema, filme, lugar):
    cond = False
    for sala in cinema:
        if filme == sala[2]:
            if lugar not in sala [1]:
                cond = True
    return cond

def vende_bilhete (cinema, filme, lugar):
    for sala in cinema:
        if filme == sala [2]:
            if lugar not in sala [1]:
                sala[1].append (lugar)
    return cinema

def listar_disponibilidades (cinema):
    for sala in cinema:
        print (sala [2], sala[0] - len (sala[1]))
    return cinema

def inserir_sala (cinema, sala):
    cinema.append (sala)
    return cinema

def menu():
    print ("""
1- Listar Filmes em exibição
2- Verificar disponibilidade
3- Listar Cinema
4- Comprar Bilhete
5- Listar lugares disponíveis
6- Inserir Sala
0- Terminar""")
```
